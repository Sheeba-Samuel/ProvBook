```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = np.linspace(0.0, 4.0)
y1 = np.cos(4 * np.pi * x1) * np.exp(-x1)
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.ylabel('Damped oscillation')
plt.show()
```

```python
x2 = np.linspace(0.0, 3.0)
y2 = np.cos(2 * np.pi * x2)
```

```python
plt.subplot(3, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
plt.show()
```
